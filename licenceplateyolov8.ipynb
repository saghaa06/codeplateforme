```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 model (e.g. small version)
model = YOLO("yolov8n.pt")  # n = nano, s = small, m = medium, l = large

# Train the model
model.train(
    data="/content/License-Plate-Detection-1/data.yaml",  # path to YAML
    epochs=50,        # number of epochs (increase if needed)
    imgsz=640,        # image size
    batch=16,         # adjust batch size based on your GPU
    name="plate_detector"
)
```

```python
# Evaluate the model on the test set
metrics = model.val(data="/content/License-Plate-Detection-1/data.yaml") # Specify the path to your data.yaml

# Print the evaluation metrics
print(metrics)
```

```python
# Make predictions on a new image
# Replace "path/to/your/image.jpg" with the actual path to your image
results = model.predict(source="/content/try2.jpg", save=True) # Set save=True to save the predicted images

# Display the image with predictions
from IPython.display import Image, display

# The results are saved in runs/detect/predict by default
# You might need to adjust the path if you specified a different save directory during prediction
predicted_image_path = results[0].save_dir + '/car5.jpg' # Assuming only one image was predicted and correcting the filename

display(Image(filename=predicted_image_path))
```
